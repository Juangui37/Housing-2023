# Creating a scatterplot for all my numeric variables that have a correlation of at least .5 

```python
# Drop non-numeric columns
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Correlation matrix of numeric variables
corr = df_numeric.corr()

# Set threshold for correlation value
corr_threshold = 0.5

# Filter for variables with correlation greater than threshold
high_corr_vars = corr[abs(corr) > corr_threshold].stack().reset_index()
high_corr_vars.columns = ['var1', 'var2', 'corr']
high_corr_vars = high_corr_vars.loc[high_corr_vars['var1'] != high_corr_vars['var2']]

# Scatter plots for variables with high correlation
for i in range(len(high_corr_vars)):
    var1 = high_corr_vars.iloc[i]['var1']
    var2 = high_corr_vars.iloc[i]['var2']
    corr_value = high_corr_vars.iloc[i]['corr']
    plt.figure()
    plt.scatter(df[var1], df[var2])
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.title(f'{var1} vs {var2} (Corr: {corr_value:.2f})')
    plt.show()
```

# Installing the dependicies for creating a machine learning CART model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeRegressor
```

Creating my Target variables

```python
df_filtered = df[(df['ListPrice'] >= 240000) & (df['ListPrice'] <= 290000)]
```

```python
# Split the data into features and target variable
X = df_filtered.drop('ListPrice', axis=1)
y = df_filtered['ListPrice']
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Remove categorical variables from the feature set
categorical_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']
X_train = X_train.drop(categorical_cols, axis=1)
X_val = X_val.drop(categorical_cols, axis=1)
```

```python
# Create one-hot encoder object
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
# Encode the categorical variables in the training data
X_train_encoded = encoder.fit_transform(X_train)
```

```python
# Transform the validation data using the trained encoder
X_val_encoded = encoder.transform(X_val)
```

```python
# Train the decision tree model on the encoded training data
model = DecisionTreeRegressor(random_state=0)
model.fit(X_train_encoded, y_train)
```

```python
# Evaluate the model on the training and validation data
train_score = model.score(X_train_encoded, y_train)
val_score = model.score(X_val_encoded, y_val)
print(f"Training score: {train_score:.3f}")
print(f"Validation score: {val_score:.3f}")
```

```python
from sklearn.tree import plot_tree

# Visualize the CART model
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X_train.columns, filled=True)
plt.show()
```

```python
# Select the columns used to train the model
cols_used = encoder.get_feature_names(input_features=categorical_cols)
X_train_encoded_df = pd.DataFrame(X_train_encoded.toarray(), columns=cols_used)

# Visualize the CART model
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=X_train_encoded_df.columns, filled=True)
plt.show()
```

# Now lets create the regression model where we can input specific values for a model created by AIC and get back a price of a home from this dataset

```python
import statsmodels.api as sm
import itertools

# Define your dependent variable and independent variables
y = df['ListPrice']
X = df.drop(['ListPrice'], axis=1)

# Define a function to calculate AIC for a given set of variables
def calculate_aic(X, y, feature_set):
    model = sm.OLS(y, X[list(feature_set)]).fit()
    aic = model.aic
    return (aic, feature_set)

# Create a list of all possible combinations of independent variables
all_features = X.columns
possible_feature_sets = []
for i in range(1, len(all_features)+1):
    possible_feature_sets += itertools.combinations(all_features, i)

# Calculate AIC for each possible feature set
results = []
for feature_set in possible_feature_sets:
    results.append(calculate_aic(X, y, feature_set))

# Sort the results by AIC value
sorted_results = sorted(results)

# Print the best result
best_result = sorted_results[0]
print('AIC:', best_result[0])
print('Features:', best_result[1])
```
