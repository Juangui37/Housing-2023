```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file with query results
MySQLresults_df = pd.read_excel("MongoDB_query_reults.xlsx")

# Loop through each query and create a bar chart
for i, row in MySQLresults_df.iterrows():
    query = row["Query"]
    result = row["Result"]
    plt.bar(query, result)
    plt.title(f"Query {i+1}")
    plt.xlabel("Query")
    plt.ylabel("Execution Time (seconds)")
    plt.text(query, result, f"{result:.15f}", ha="center", va="bottom")
    plt.show()
```
